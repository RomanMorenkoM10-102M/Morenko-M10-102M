Посчитать среднюю цену закрытия акций Tesla, в периуд с открытия по 10.01.2012

```python
df.index = df.Date
df.loc[:"2012-01-10"].Close.median()
```

Вывести график цены закрытия

```python
from matplotlib.pyplot import figure, show
figure(figsize=(15, 3), dpi=120)
df['Close'].plot()
show()
```

Примените эту формулу к столбцу Open (используйте метод apply).

```python
xmin = df.Open.min()
xmax = df.Open.max()
df['Open'].apply(lambda x: (x-xmin)/(xmax-xmin))
```

Что такое SMA (Simple Moving Average) с окном 12,15,30. Можно посмотреть внешние решения:

```python
df.Close.rolling(window=12).mean().plot()
df.Close.rolling(window=15).mean().plot()
df.Close.rolling(window=30).mean().plot()
show()
```
